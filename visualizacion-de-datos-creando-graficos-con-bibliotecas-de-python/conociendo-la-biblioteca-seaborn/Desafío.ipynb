````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
url = 'inmigrantes_canada.csv'

inmigrantes = pd.read_csv(url)
````

````python
paises = ['Brasil', 'Argentina', 'Perú', 'Colombia']
anos = [str(ano) for ano in range(1980,2014)]
inmigrantes_wide = pd.melt(inmigrantes, value_vars=anos, id_vars="Pais", var_name="año", value_name="inmigrantes")
inmigrantes_wide_latinoamerica = inmigrantes_wide[inmigrantes_wide["Pais"].isin(paises)].reset_index(drop=True)
inmigrantes_wide_latinoamerica["año"] = inmigrantes_wide_latinoamerica["año"].astype(int)
inmigrantes_wide_latinoamerica
````

````python
sns.set_theme(style="white")
def generar_comparacion(colores):
    fig, ax = plt.subplots(figsize=(10,8))
    sns.
    sns.color_palette(palette=colores)
    ax = sns.lineplot(inmigrantes_wide_latinoamerica, x = 'año', y= 'inmigrantes', markers=True, hue='Pais',lw=3)
    ax.set_xlabel("Años", fontsize=20)
    ax.set_ylabel('Nro. de Inmigrantes', fontsize=20)
    ax.legend()
    plt.suptitle("Nro. de Inmigrantes por año \n en algunos paises de America del Norte",fontsize=30)
    plt.show()
````

````python
import pandas as pd
import matplotlib.pyplot as plt
````

````python
tiendas = ['A', 'B', 'C', 'D']
ventas_2022 = {'Ene': [100, 80, 150, 50],
'Feb': [120, 90, 170, 60],
'Mar': [150, 100, 200, 80],
'Abr': [180, 110, 230, 90],
'May': [220, 190, 350, 200],
'Jun': [230, 150, 280, 120],
'Jul': [250, 170, 300, 140],
'Ago': [260, 180, 310, 150],
'Sep': [240, 160, 290, 130],
'Oct': [220, 140, 270, 110],
'Nov': [400, 220, 350, 190],
'Dec': [300, 350, 400, 250]}
meses = list(ventas_2022.keys())
````

````python
ventas = pd.DataFrame(ventas_2022, index=tiendas)
ventas
````

````python
fig, axs = plt.subplots(2,2,figsize=(10, 12))
tiendas = ['A','B','C','D']
colores = ['green','blue','red','cyan']
for i, ax in enumerate(axs.ravel()):
      tienda = tiendas[i]
      ax.plot(meses, ventas.loc[tienda], marker='o', color=colores[i], alpha=0.5, lw=3)
      ax.set_title(f'Tienda {tienda}')
      ax.set_xlabel('Meses')
      ax.set_ylabel('Ventas')
      ax.set_ylim(0,500)
      ax.grid()

fig.subplots_adjust()
fig.suptitle('Ventas Mensuales de las Tiendas en 2022', fontsize=16)
plt.tight_layout(rect=(0.05, 0.05, 0.9, 0.95))
plt.show()
````

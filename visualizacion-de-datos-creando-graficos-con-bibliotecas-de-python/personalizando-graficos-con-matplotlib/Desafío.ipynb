````python
import pandas as pd
import matplotlib.pyplot as plt
````

````python
tiendas = ['A', 'B', 'C', 'D']
ventas_2022 = {'Ene': [100, 80, 150, 50],
'Feb': [120, 90, 170, 60],
'Mar': [150, 100, 200, 80],
'Abr': [180, 110, 230, 90],
'May': [220, 190, 350, 200],
'Jun': [230, 150, 280, 120],
'Jul': [250, 170, 300, 140],
'Ago': [260, 180, 310, 150],
'Sep': [240, 160, 290, 130],
'Oct': [220, 140, 270, 110],
'Nov': [400, 220, 350, 190],
'Dec': [300, 350, 400, 250]}
meses = list(ventas_2022.keys())
````

````python
ventas = pd.DataFrame(ventas_2022, index=tiendas)
ventas
````

````python
plt.style.use('fivethirtyeight')
````

````python
fig, axs = plt.subplots(2,2, figsize=(12,8))
colores = ['blue', 'orange', 'green', 'red']
markers = ['o', 's', 'd', '*']
for i, ax in enumerate(axs.flat):
      x = meses
      y = ventas.iloc[i]
      tienda = tiendas[i]
      ax.plot(x, y,color=colores[i], marker=markers[i], lw=3, alpha=0.6, label = f'Tienda {tienda}', markersize=5)
      ax.set_title(f'Venta Mensual: Tienda {tienda}', fontsize = 14, loc = "left", pad = 10)
      ax.set_ylabel("Ventas", fontsize=10)
      ax.set_xlabel("Meses", fontsize=10)
      ax.tick_params(axis="both",labelsize=8)
      ax.annotate('Mayor venta', xy=(y.idxmax(), y.max()), xytext=(0, 20), textcoords='offset points', arrowprops=dict(arrowstyle='->', lw=0.5), ha='center')
      ax.legend()
      ax.grid(linestyle="--")

fig.tight_layout(h_pad=2, rect=(0.05,0.1,0.9,0.9))
fig.suptitle('Análisis Comparativo de Ventas por Sucursal - 2022', fontsize=15, fontweight="bold")
plt.show()
````

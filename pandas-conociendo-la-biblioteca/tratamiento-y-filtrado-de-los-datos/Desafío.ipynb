```python
import pandas as pd

url = "https://gist.githubusercontent.com/ahcamachod/807a2c1cf6c19108b2b701ea1791ab45/raw/fb84f8b2d8917a89de26679eccdbc8f9c1d2e933/alumnos.csv"

df = pd.read_csv(url)
```

```python
print(f"Los valores nulos en las columnas son: \n{df.isnull().sum()}")

# Se encontraron valores nulos en Nota
"""
sin_nota = df[df["Nota"].isnull()]["Nombre"].tolist()
for nombre in sin_nota:
      print(df[df["Nombre"]==nombre])
"""
# Como se puede ver no se puede deducir nada pues los datos son unicos entonces los quitaremos
df_modificado = df.dropna().reset_index(drop=True)
df_new = df_modificado[~df_modificado.isin(["Alicia","Carlos"])].dropna().reset_index(drop=True)

df_aprobado = df_new[df_new["Aprobado"].isin(["True","Verdadero"])]

print(df_aprobado)

df_aprobado["Nota"] = df_aprobado["Nota"].replace(7.0,8.0)

print(df_aprobado)

df_aprobado.to_csv("alumnos_aprovados.csv", index=False)
```

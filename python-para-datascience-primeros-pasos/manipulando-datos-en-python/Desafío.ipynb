8 - Crea un programa que solicite dos valores numéricos, un numerador y un denominador, y realice la división entre los dos valores. Asegúrate de que el valor del denominador no sea igual a 0.

```python
num1, num2 = input("Ingresa el numerador y el denominador separados por espacio: ").split(" ")
if int(num2) != 0:
    division = int(num1) / int(num2)
    print("El resultado de la división es:", division)
else:
    print("Error: El denominador no puede ser igual a 0.")
```

9 - Crea un programa que solicite dos valores numéricos, un operador y una potencia, y realice la exponenciación entre estos dos valores.

```python
num1, num2, operador, potencia = input("Ingresa dos números, un operador (+, -, *, /) y una potencia separados por espacio: ").split(" ")
num1 = int(num1)
num2 = int(num2)
potencia = int(potencia)
resultado = None
if operador == "+":
    resultado = num1 + num2
elif operador == "-":
    resultado = num1 - num2
elif operador == "*":
    resultado = num1 * num2
elif operador == "/":
    if num2 != 0:
        resultado = num1 / num2
    else:
        print("Error: El denominador no puede ser igual a 0.")
        resultado = None

if resultado is not None:
    print("El resultado de la operación es:", resultado)

exponenciacion = resultado ** potencia
print("El resultado de la exponenciación es:", exponenciacion)
```

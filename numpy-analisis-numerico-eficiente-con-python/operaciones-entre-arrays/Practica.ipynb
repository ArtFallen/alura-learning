Continuando con el proyecto de las naranjas/toronjas, ahora debes calcular el coeficiente angular y lineal para la recta de las naranjas y para la recta de las toronjas. Utiliza la fórmula de mínimos cuadrados para encontrar cada uno.

---

```python
import pandas as pd
from numpy import loadtxt, arange
import matplotlib.pyplot as plt
```

```python
url = "https://gist.githubusercontent.com/ahcamachod/9be09de793dc3bf1e6c3d98eb4e5b1ef/raw/21b85572693200040e11284ef6dcfc3457ec8e11/citrus.csv"

df = pd.read_csv(url)
print(df.shape)

# Conociendo la cantidad de filas podemos hacer un rango
datos = loadtxt(url, delimiter=",",usecols= arange(1,df.shape[1]),skiprows=1)
```

```python
naranjas = datos[:5000]
toronjas = datos[5000:]

naranjas_diametro = naranjas[:,0]
naranjas_peso = naranjas[:,1]

toronjas_diametro = toronjas[:,0]
toronjas_peso = toronjas[:,1]
```

```python
n = naranjas.shape[0]
X = naranjas_diametro
Y = naranjas_peso
m = (n * sum(X * Y) - sum(X)*sum(Y)) / (n * sum(X**2) - sum(X)**2)
b = (sum(Y)- m * sum(X))/ n
y = m * naranjas_diametro + b
```

```python
plt.plot(X, Y)
plt.plot(X, y)
plt.show()
```

```python
n = toronjas.shape[0]
X = toronjas_diametro
Y = toronjas_peso
m = (n * sum(X * Y) - sum(X)*sum(Y))/ (n * sum(X**2) - (sum(X))**2)
b = (sum(Y)- m * sum(X)) / n
y = m * X + b
```

```python
plt.cla()
plt.plot(X, Y)
plt.plot(X, y)
plt.show()
```

Continuando con el proyecto de las naranjas/toronjas, ahora debes calcular el coeficiente angular utilizando la generación de números aleatorios. Supongamos que ya conoces el valor de b y que este es igual a 17.

---

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = "https://gist.githubusercontent.com/ahcamachod/9be09de793dc3bf1e6c3d98eb4e5b1ef/raw/21b85572693200040e11284ef6dcfc3457ec8e11/citrus.csv"

df = pd.read_csv(url)
print(df.shape)

# Conociendo la cantidad de filas podemos hacer un rango
datos = np.loadtxt(url, delimiter=",",usecols= np.arange(1,df.shape[1]),skiprows=1)
```

```python
naranjas = datos[:5000]
toronjas = datos[5000:]

naranjas_diametro = naranjas[:,0]
naranjas_peso = naranjas[:,1]

toronjas_diametro = toronjas[:,0]
toronjas_peso = toronjas[:,1]
```

```python
n = naranjas.shape[0]
X = naranjas_diametro
Y = naranjas_peso
norma = np.array([])
np.random.seed(30)
coeficientes_angulares = np.random.uniform(-30,30,n)
b = 17
for coeficiente in coeficientes_angulares:
      y_pred = X*coeficiente + b
      norma = np.append(norma, np.linalg.norm(Y-y_pred))

indice_mejor = np.argmin(norma)
m = coeficientes_angulares[indice_mejor]
y = m * X + b
```

```python
plt.plot(X, Y)
plt.plot(X, y)
plt.show()
```

```python
n = toronjas.shape[0]
X = toronjas_diametro
Y = toronjas_peso
norma = np.array([])
np.random.seed(21)
coeficientes_angulares = np.random.uniform(-30,30,n)
b = 17
for coeficiente in range(0,n):
      y_pred = X*coeficiente + b
      norma = np.append(norma,np.linalg.norm(Y-y_pred))

indice_mejor = np.argmin(norma)
m = coeficientes_angulares[indice_mejor]
y = m * X + b
```

```python
plt.cla()
plt.plot(X, Y)
plt.plot(X, y)
plt.show()
```

```python
import pandas as pd
```

```python
url = "1-SEEG10_GERAL-BR_UF_2022.10.27-FINAL-SITE.xlsx"

emision_gases = pd.read_excel(url , sheet_name = 'GEE Estados')
```

```python
emision_gases.head()
```

```python
años = list(emision_gases.loc[:,1970:2021])
columnas = list(emision_gases.loc[:,'Nível 1 - Setor':'Produto'])
```

```python
emision_gases_por_año = pd.melt(emision_gases, id_vars=columnas, value_vars=años, value_name='Emision', var_name='Año')
```

```python
emision_estado_sector = emision_gases_por_año.groupby(['Estado', 'Nível 1 - Setor'])['Emision'].sum()
```

```python
emision_estado_sector.xs('Energia', level = 1)
```

```python
emision_estado_sector.xs('MG', level = 0).idxmax()
```

```python
emision_estado_sector.groupby(level = 0).idxmax()
```

```python
emision_estado_sector.groupby(level=1).idxmax()
```
